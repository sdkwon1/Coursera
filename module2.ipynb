```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('data.csv')

for col in df.select_dtypes(include=['object']).columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

#   
print(df.dtypes)
df.head()

df.corr()

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
#    
selected_columns = ['incomeperperson', 'breastcancerper100th', 'internetuserate']
df_selected = df[selected_columns]

#       
for column in df_selected.columns:
    print(f"### {column}   ###")
    print(df_selected[column].value_counts(dropna=False))  # NaN    
    print("\n")

#   , ,   
for column in df_selected.columns:
    print(f"### {column}   ###")
    print(f": {df_selected[column].min()}")
    print(f": {df_selected[column].max()}")
    print(f": {df_selected[column].median()}")
    print(f": {df_selected[column].mean()}")
    print(f"  : {df_selected[column].isnull().sum()}")
    print(f"  : {df_selected[column].isnull().mean() * 100:.2f}%")
    print("\n")
```
